Random Name Generator for the EastCoastStates: Florida, Georgia, Maine, Maryland, Massachusetts, New York, New Jersey, North
Carolina, Pennsylvania, South Carolina and Virginia.


Generator loads the names from the text files of each state consisting first names of both males and females ,and last names.


```python
import random
```

EastCoastNames class that will load up all the first and last names by each state and gender during its initialization. 

```python
# class to loadnames, generate first names, last names, full names
class EastCoastNames:
    #constructor 
    def __init__(self):
        self.first_names = {}
        self.last_names = {}
        self.load_names()
    
    # method to load names from the files to dictionaries
    def load_names(self):
        states = ["Florida", "Georgia", "Maine", "Maryland", "Massachusetts", "New York", "New Jersey", "North Carolina", "Pennsylvania", "South Carolina", "Virginia"]
        for state in states:
            self.first_names[state] = {"Male": [], "Female": []}
            self.last_names[state] = []
            with open(f"EastCoastStates/{state}.txt", "r") as f:
                for line in f:
                    #print(line)
                    #print(len(line.split(",")))
                    if(len(line.split(","))==2):
                        name, gender = line.strip().split(",")
                        #print("In if", name)
                        #print(type(name))
                        if gender == "Male":
                            self.first_names[state]["Male"].append(name)
                        else:
                            self.first_names[state]["Female"].append(name)
                    else:
                        name = line.strip().split(",")
                        #print(type(name))
                        #print("IN else", name)
                        self.last_names[state].append(name[0])
                #print("First names:",self.first_names)
                #print("Last names:",self.last_names)
            #print(self.first_names)
            #print(self.last_names)
    
    # method to generate the firstnames using state and gender as parameters
    def first_name(self, state, gender):
        if state is None:
            state = random.choice(list(self.first_names.keys()))
        if gender is None:
            gender = random.choice(["Male", "Female"])
        #print("In first_name",state,gender)
        return random.choice(self.first_names[state][gender]), gender, state
    
    # method to generate lastnames using state as parameter
    def last_name(self, state=None):
        if state is None:
            state = random.choice(list(self.last_names.keys()))
        #print("In last_name",state,gender)
        return random.choice(self.last_names[state]), state
    
    # method to generate complete fullnames using firstname and lastname methods
    def full_name(self, state, gender):
        if state is None:
            state = random.choice(list(self.first_names.keys()))
        if gender is None:
            gender = random.choice(["Male", "Female"])
        first, gender, state = self.first_name(state, gender)
        last, state = self.last_name(state)
        fullname = first+ " "+last
        #print("In fullname:",first,last,state)
        return fullname, gender, state

```

## Task 2:

Roman Naming Convention follows below format:
            
            <praenomen> <nomen gentile> <cognomen>  
 praenomen (or given name), nomen gentile (or Gens name) and cognomen
    
    Praenomen represents personal name chosen by the parents of a Roman child.
    Nomen Gentile which may be translated as "race", "family", or "clan", constituted an extended Roman family
    cognomen 
   



```python
import random
```

RomanNamingConvention class for creating the names of Roman culture

```python
class RomanNamingConvention:
    def __init__(self):
        # Initialize instance variables to empty dictionaries and an empty list
        self.praenomen = {}
        self.cognomen = {}
        self.nomen_gentile = []
        
        # Call the load_names method to populate the instance variables
        self.load_names()
    
    def load_names(self):
        # Initialize empty lists for male and female praenomen names
        self.praenomen = {"Male": [], "Female": []}
        
        # Initialize empty lists for male and female cognomen names
        self.cognomen = {"Male": [], "Female": []}
        
        # Initialize an empty list for nomen gentile names
        self.nomen_gentile = []
        
        # Open and read the file "praenomen.txt"
        with open("RNC/praenomen.txt", "r") as f:
            # Loop through each line of the file
            for line in f:
                #print(line)
                #print(len(line.split(",")))
                # Split each line by comma and assign the name and gender to variables
                name, gender = line.strip().split(",")
                
                #print("In if", name)
                #print(type(name))
                # If the gender is male, append the name to the male praenomen list
                if gender == "Male":
                    self.praenomen["Male"].append(name)
                # If the gender is female, append the name to the female praenomen list
                else:
                    self.praenomen["Female"].append(name)
        
        # Open and read the file "cognomen.txt"
        with open("RNC/cognomen.txt", "r") as f:
            # Loop through each line of the file
            for line in f:
                #print(line)
                #print(len(line.split(",")))
                # Split each line by comma and assign the name and gender to variables
                name, gender = line.strip().split(",")
                #print("In if", name)
                #print(type(name))
                # If the gender is male, append the name to the male cognomen list
                if gender == "Male":
                    self.cognomen["Male"].append(name)
                # If the gender is female, append the name to the female cognomen list
                else:
                    self.cognomen["Female"].append(name)
                
        # Open and read the file "nomen_gentile.txt"
        with open("RNC/nomen_gentile.txt", "r") as f:
            # Loop through each line of the file
            for line in f:
                #print(line)
                #print(len(line.split(",")))
                # Split each line by comma and assign the name to a variable
                name = line.strip().split(",")
                
                #print(type(name))
                #print("IN else", name)
                # Append the name to the nomen gentile list
                self.nomen_gentile.append(name[0])
                
                # The nomen gentile file only has one name per line, so there's no gender to worry about
```
